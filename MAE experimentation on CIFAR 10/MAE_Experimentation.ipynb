```python
# ============================================================
# Masked Autoencoder (MAE) Experimentation on CIFAR-10
# Varies: masking ratio, patch size, encoder depth/width
# Pure PyTorch — paste directly into Google Colab
# ============================================================

# ── 0. Install / imports ─────────────────────────────────────
!pip install einops timm --quiet

import math, random, itertools, time
from dataclasses import dataclass, field
from typing import List

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from einops import rearrange

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {DEVICE}")


# ── 1. Config dataclass ──────────────────────────────────────
@dataclass
class MAEConfig:
    # Image
    img_size:        int   = 32          # CIFAR-10 native size
    in_channels:     int   = 3

    # Patch
    patch_size:      int   = 4           # 4 → 64 patches; 8 → 16 patches

    # Masking
    mask_ratio:      float = 0.75

    # Encoder (ViT backbone)
    enc_dim:         int   = 192
    enc_depth:       int   = 6
    enc_heads:       int   = 3

    # Decoder (lightweight)
    dec_dim:         int   = 96
    dec_depth:       int   = 2
    dec_heads:       int   = 3

    # Training
    epochs:          int   = 50
    batch_size:      int   = 256
    lr:              float = 1.5e-4
    weight_decay:    float = 0.05
    warmup_epochs:   int   = 5
    label:           str   = "default"   # human-readable run name

    @property
    def num_patches(self):
        return (self.img_size // self.patch_size) ** 2

    @property
    def patch_dim(self):
        return self.in_channels * self.patch_size ** 2


# ── 2. Transformer building blocks ──────────────────────────
class MultiHeadSelfAttention(nn.Module):
    def __init__(self, dim, heads, dropout=0.0):
        super().__init__()
        assert dim % heads == 0
        self.heads, self.scale = heads, (dim // heads) ** -0.5
        self.qkv  = nn.Linear(dim, dim * 3, bias=False)
        self.proj = nn.Linear(dim, dim)
        self.drop = nn.Dropout(dropout)

    def forward(self, x):
        B, N, C = x.shape
        qkv = self.qkv(x).reshape(B, N, 3, self.heads, C // self.heads).permute(2,0,3,1,4)
        q, k, v = qkv.unbind(0)
        attn = (q @ k.transpose(-2,-1)) * self.scale
        attn = self.drop(attn.softmax(-1))
        x = (attn @ v).transpose(1,2).reshape(B, N, C)
        return self.proj(x)


class TransformerBlock(nn.Module):
    def __init__(self, dim, heads, mlp_ratio=4.0, dropout=0.0):
        super().__init__()
        self.norm1 = nn.LayerNorm(dim)
        self.attn  = MultiHeadSelfAttention(dim, heads, dropout)
        self.norm2 = nn.LayerNorm(dim)
        hidden     = int(dim * mlp_ratio)
        self.mlp   = nn.Sequential(
            nn.Linear(dim, hidden), nn.GELU(), nn.Dropout(dropout),
            nn.Linear(hidden, dim), nn.Dropout(dropout))

    def forward(self, x):
        x = x + self.attn(self.norm1(x))
        x = x + self.mlp(self.norm2(x))
        return x


def sincos_pos_embed(num_patches, dim):
    """Fixed 2D sine-cosine positional embedding (no CLS)."""
    grid = int(num_patches ** 0.5)
    assert grid * grid == num_patches
    pos  = torch.arange(grid, dtype=torch.float)
    gy, gx = torch.meshgrid(pos, pos, indexing='ij')   # (grid, grid)
    omega  = torch.arange(dim // 4, dtype=torch.float) / (dim // 4)
    omega  = 1. / (10000 ** omega)                      # (dim//4,)
    def enc(g):                                          # g: (grid,grid)
        g = g.flatten()[:,None] * omega[None,:]          # (N, dim//4)
        return torch.cat([g.sin(), g.cos()], -1)         # (N, dim//2)
    emb = torch.cat([enc(gy), enc(gx)], -1)             # (N, dim)
    return emb  # (num_patches, dim)


# ── 3. MAE model ─────────────────────────────────────────────
class MAE(nn.Module):
    def __init__(self, cfg: MAEConfig):
        super().__init__()
        self.cfg = cfg
        P, D = cfg.patch_size, cfg.enc_dim

        # ── Patchify projection
        self.patch_embed = nn.Linear(cfg.patch_dim, D)

        # ── Positional embeddings (fixed, no gradient)
        enc_pos = sincos_pos_embed(cfg.num_patches, D)
        dec_pos = sincos_pos_embed(cfg.num_patches, cfg.dec_dim)
        self.register_buffer("enc_pos", enc_pos)
        self.register_buffer("dec_pos", dec_pos)

        # ── Mask token (learnable)
        self.mask_token = nn.Parameter(torch.zeros(1, 1, cfg.dec_dim))
        nn.init.normal_(self.mask_token, std=0.02)

        # ── Encoder
        self.encoder_blocks = nn.ModuleList([
            TransformerBlock(D, cfg.enc_heads) for _ in range(cfg.enc_depth)])
        self.enc_norm = nn.LayerNorm(D)

        # ── Encoder→Decoder projection
        self.enc_to_dec = nn.Linear(D, cfg.dec_dim, bias=True)

        # ── Decoder
        self.decoder_blocks = nn.ModuleList([
            TransformerBlock(cfg.dec_dim, cfg.dec_heads) for _ in range(cfg.dec_depth)])
        self.dec_norm = nn.LayerNorm(cfg.dec_dim)

        # ── Pixel prediction head
        self.pred_head = nn.Linear(cfg.dec_dim, cfg.patch_dim)

        self.apply(self._init_weights)

    def _init_weights(self, m):
        if isinstance(m, nn.Linear):
            nn.init.xavier_uniform_(m.weight)
            if m.bias is not None: nn.init.zeros_(m.bias)
        elif isinstance(m, nn.LayerNorm):
            nn.init.ones_(m.weight); nn.init.zeros_(m.bias)

    def patchify(self, imgs):
        """(B,C,H,W) → (B, N, patch_dim)"""
        p = self.cfg.patch_size
        x = rearrange(imgs, 'b c (h p1) (w p2) -> b (h w) (p1 p2 c)', p1=p, p2=p)
        return x

    def unpatchify(self, x):
        """(B, N, patch_dim) → (B,C,H,W)"""
        p  = self.cfg.patch_size
        g  = int(x.shape[1] ** 0.5)
        imgs = rearrange(x, 'b (h w) (p1 p2 c) -> b c (h p1) (w p2)',
                         h=g, w=g, p1=p, p2=p, c=self.cfg.in_channels)
        return imgs

    def random_mask(self, x):
        """Returns visible tokens, mask indices, restore indices."""
        B, N, _ = x.shape
        num_mask = int(N * self.cfg.mask_ratio)
        noise    = torch.rand(B, N, device=x.device)
        ids_sort = noise.argsort(dim=1)
        ids_restore = ids_sort.argsort(dim=1)
        ids_keep    = ids_sort[:, :N - num_mask]
        x_vis       = torch.gather(x, 1, ids_keep.unsqueeze(-1).expand(-1,-1,x.shape[-1]))
        mask        = torch.ones(B, N, device=x.device)
        mask.scatter_(1, ids_keep, 0)      # 1 = masked
        return x_vis, mask, ids_restore

    def encode(self, imgs):
        x = self.patchify(imgs)                     # (B,N,patch_dim)
        x = self.patch_embed(x)                     # (B,N,enc_dim)
        x = x + self.enc_pos                        # add pos embed
        x_vis, mask, ids_restore = self.random_mask(x)
        for blk in self.encoder_blocks:
            x_vis = blk(x_vis)
        x_vis = self.enc_norm(x_vis)
        return x_vis, mask, ids_restore

    def decode(self, x_vis, ids_restore):
        B      = x_vis.shape[0]
        N      = self.cfg.num_patches
        N_vis  = x_vis.shape[1]
        N_mask = N - N_vis

        x_vis = self.enc_to_dec(x_vis)             # project to dec_dim

        # Expand mask tokens
        mask_tokens = self.mask_token.expand(B, N_mask, -1)

        # Unshuffle: put visible + mask tokens back in original order
        x_full = torch.cat([x_vis, mask_tokens], dim=1)
        x_full = torch.gather(
            x_full, 1,
            ids_restore.unsqueeze(-1).expand(-1,-1,self.cfg.dec_dim))

        x_full = x_full + self.dec_pos             # add pos embed

        for blk in self.decoder_blocks:
            x_full = blk(x_full)
        x_full = self.dec_norm(x_full)
        pred   = self.pred_head(x_full)            # (B,N,patch_dim)
        return pred

    def forward(self, imgs):
        x_vis, mask, ids_restore = self.encode(imgs)
        pred = self.decode(x_vis, ids_restore)
        loss = self.reconstruction_loss(imgs, pred, mask)
        return loss, pred, mask

    def reconstruction_loss(self, imgs, pred, mask):
        target = self.patchify(imgs)
        # Normalize per patch (MAE paper default)
        mean   = target.mean(-1, keepdim=True)
        var    = target.var(-1, keepdim=True)
        target = (target - mean) / (var + 1e-6).sqrt()
        loss   = (pred - target) ** 2
        loss   = loss.mean(-1)                      # (B, N)
        loss   = (loss * mask).sum() / mask.sum()   # only masked patches
        return loss


# ── 4. Data ──────────────────────────────────────────────────
def get_cifar10_loaders(batch_size=256, img_size=32):
    mean = (0.4914, 0.4822, 0.4465)
    std  = (0.2023, 0.1994, 0.2010)
    tf_train = transforms.Compose([
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize(mean, std)])
    tf_test  = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize(mean, std)])
    train_ds = datasets.CIFAR10(root='./data', train=True,  download=True, transform=tf_train)
    test_ds  = datasets.CIFAR10(root='./data', train=False, download=True, transform=tf_test)
    train_dl = DataLoader(train_ds, batch_size=batch_size, shuffle=True,  num_workers=2, pin_memory=True)
    test_dl  = DataLoader(test_ds,  batch_size=batch_size, shuffle=False, num_workers=2, pin_memory=True)
    return train_dl, test_dl


# ── 5. Scheduler ─────────────────────────────────────────────
def cosine_schedule_with_warmup(optimizer, warmup_epochs, total_epochs, base_lr):
    def lr_lambda(epoch):
        if epoch < warmup_epochs:
            return epoch / max(1, warmup_epochs)
        progress = (epoch - warmup_epochs) / max(1, total_epochs - warmup_epochs)
        return 0.5 * (1 + math.cos(math.pi * progress))
    return torch.optim.lr_scheduler.LambdaLR(optimizer, lr_lambda)


# ── 6. Training loop ─────────────────────────────────────────
def train_mae(cfg: MAEConfig, train_dl, verbose=True):
    model     = MAE(cfg).to(DEVICE)
    optimizer = torch.optim.AdamW(model.parameters(), lr=cfg.lr,
                                  weight_decay=cfg.weight_decay, betas=(0.9, 0.95))
    scheduler = cosine_schedule_with_warmup(optimizer, cfg.warmup_epochs, cfg.epochs, cfg.lr)

    history = []
    t0 = time.time()
    for epoch in range(1, cfg.epochs + 1):
        model.train()
        epoch_loss = 0.0
        for imgs, _ in train_dl:
            imgs = imgs.to(DEVICE)
            loss, _, _ = model(imgs)
            optimizer.zero_grad()
            loss.backward()
            nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()
            epoch_loss += loss.item() * imgs.size(0)
        scheduler.step()
        avg = epoch_loss / len(train_dl.dataset)
        history.append(avg)
        if verbose and (epoch % 10 == 0 or epoch == 1):
            elapsed = time.time() - t0
            print(f"[{cfg.label}] Epoch {epoch:3d}/{cfg.epochs}  loss={avg:.4f}  "
                  f"lr={scheduler.get_last_lr()[0]:.2e}  ({elapsed:.0f}s)")
    return model, history


# ── 7. Visualisation helpers ──────────────────────────────────
CIFAR_MEAN = torch.tensor([0.4914, 0.4822, 0.4465])
CIFAR_STD  = torch.tensor([0.2023, 0.1994, 0.2010])

def denorm(t):
    """(C,H,W) tensor → (H,W,C) numpy in [0,1]"""
    t = t.cpu() * CIFAR_STD[:,None,None] + CIFAR_MEAN[:,None,None]
    return t.clamp(0,1).permute(1,2,0).numpy()


def visualize_reconstructions(model, test_dl, cfg, n=6, title=""):
    model.eval()
    imgs, _ = next(iter(test_dl))
    imgs    = imgs[:n].to(DEVICE)
    with torch.no_grad():
        _, pred, mask = model(imgs)

    p   = cfg.patch_size
    g   = cfg.img_size // p

    fig, axes = plt.subplots(3, n, figsize=(n * 2.2, 6.5))
    fig.suptitle(title, fontsize=13, fontweight='bold')
    col_labels = ['Original', 'Masked Input', 'Reconstruction']

    for i in range(n):
        img   = imgs[i]                                     # (C,H,W)
        # ── Masked input
        m = mask[i].reshape(g, g)                           # (g,g) float on device
        m_img = img.clone()
        for r in range(g):
            for c in range(g):
                if m[r,c] > 0.5:
                    m_img[:, r*p:(r+1)*p, c*p:(c+1)*p] = 0.5
        # ── Reconstruction (paste predicted patches onto original for masked regions)
        rec_patches = model.unpatchify(pred)[i]             # (C,H,W)
        rec_img = img.clone()
        for r in range(g):
            for c in range(g):
                if m[r,c] > 0.5:
                    rec_img[:, r*p:(r+1)*p, c*p:(c+1)*p] = \
                        rec_patches[:, r*p:(r+1)*p, c*p:(c+1)*p]

        for row, im in enumerate([denorm(img), denorm(m_img), denorm(rec_img)]):
            ax = axes[row, i]
            ax.imshow(im)
            ax.axis('off')
            if i == 0:
                ax.set_ylabel(col_labels[row], fontsize=10)
    plt.tight_layout()
    plt.show()


def plot_loss_curves(all_histories, labels, title="Training Loss Curves"):
    plt.figure(figsize=(9, 4))
    for h, lbl in zip(all_histories, labels):
        plt.plot(h, label=lbl)
    plt.xlabel("Epoch"); plt.ylabel("Reconstruction Loss (MSE)")
    plt.title(title); plt.legend(fontsize=8); plt.grid(True, alpha=0.3)
    plt.tight_layout(); plt.show()


def final_loss_bar(final_losses, labels, title="Final Reconstruction Loss"):
    plt.figure(figsize=(max(6, len(labels)*1.2), 4))
    colors = plt.cm.tab10(range(len(labels)))
    bars = plt.bar(labels, final_losses, color=colors)
    for bar, v in zip(bars, final_losses):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.001,
                 f"{v:.4f}", ha='center', va='bottom', fontsize=8)
    plt.ylabel("Final Loss"); plt.title(title)
    plt.xticks(rotation=30, ha='right', fontsize=8)
    plt.tight_layout(); plt.show()


# ── 8. Experiment grid ────────────────────────────────────────

# Base config — change these to taste
BASE = dict(img_size=32, in_channels=3, epochs=50, batch_size=256,
            lr=1.5e-4, weight_decay=0.05, warmup_epochs=5,
            enc_dim=192, enc_depth=6, enc_heads=3,
            dec_dim=96,  dec_depth=2,  dec_heads=3)

# ① Vary masking ratio  (patch_size=4, enc_dim=192)
mask_ratios = [0.25, 0.50, 0.75, 0.85]

# ② Vary patch size  (mask_ratio=0.75, enc_dim=192)
patch_sizes = [2, 4, 8]          # 2→256 patches, 4→64, 8→16

# ③ Vary encoder size  (mask_ratio=0.75, patch_size=4)
encoder_sizes = [
    dict(enc_dim=96,  enc_depth=3, enc_heads=3, label="Tiny"),
    dict(enc_dim=192, enc_depth=6, enc_heads=3, label="Small"),
    dict(enc_dim=384, enc_depth=6, enc_heads=6, label="Base"),
]

train_dl, test_dl = get_cifar10_loaders(batch_size=BASE['batch_size'])


# ─── EXPERIMENT 1: Masking Ratio ─────────────────────────────
print("\n" + "="*55)
print("EXPERIMENT 1: Varying Masking Ratio")
print("="*55)
hist_mask, labels_mask, models_mask = [], [], []
for mr in mask_ratios:
    lbl = f"mask={mr}"
    cfg = MAEConfig(**{**BASE, 'mask_ratio': mr, 'patch_size': 4, 'label': lbl})
    model, hist = train_mae(cfg, train_dl)
    hist_mask.append(hist)
    labels_mask.append(lbl)
    models_mask.append((model, cfg))

plot_loss_curves(hist_mask, labels_mask, "Exp 1 — Masking Ratio")
final_loss_bar([h[-1] for h in hist_mask], labels_mask, "Exp 1 — Final Loss by Masking Ratio")
for model, cfg in models_mask:
    visualize_reconstructions(model, test_dl, cfg, title=f"mask_ratio={cfg.mask_ratio}")


# ─── EXPERIMENT 2: Patch Size ────────────────────────────────
print("\n" + "="*55)
print("EXPERIMENT 2: Varying Patch Size")
print("="*55)
hist_patch, labels_patch, models_patch = [], [], []
for ps in patch_sizes:
    lbl = f"patch={ps}x{ps}"
    cfg = MAEConfig(**{**BASE, 'patch_size': ps, 'mask_ratio': 0.75, 'label': lbl})
    model, hist = train_mae(cfg, train_dl)
    hist_patch.append(hist)
    labels_patch.append(lbl)
    models_patch.append((model, cfg))

plot_loss_curves(hist_patch, labels_patch, "Exp 2 — Patch Size")
final_loss_bar([h[-1] for h in hist_patch], labels_patch, "Exp 2 — Final Loss by Patch Size")
for model, cfg in models_patch:
    visualize_reconstructions(model, test_dl, cfg, title=f"patch_size={cfg.patch_size}")


# ─── EXPERIMENT 3: Encoder Size ──────────────────────────────
print("\n" + "="*55)
print("EXPERIMENT 3: Varying Encoder Size")
print("="*55)
hist_enc, labels_enc, models_enc = [], [], []
for es in encoder_sizes:
    lbl = f"enc={es['label']} (d={es['enc_dim']},L={es['enc_depth']})"
    cfg = MAEConfig(**{**BASE,
                       'enc_dim':   es['enc_dim'],
                       'enc_depth': es['enc_depth'],
                       'enc_heads': es['enc_heads'],
                       'patch_size': 4, 'mask_ratio': 0.75, 'label': lbl})
    model, hist = train_mae(cfg, train_dl)
    hist_enc.append(hist)
    labels_enc.append(lbl)
    models_enc.append((model, cfg))

plot_loss_curves(hist_enc, labels_enc, "Exp 3 — Encoder Size")
final_loss_bar([h[-1] for h in hist_enc], labels_enc, "Exp 3 — Final Loss by Encoder Size")
for model, cfg in models_enc:
    visualize_reconstructions(model, test_dl, cfg,
                               title=f"enc={cfg.enc_dim}d, {cfg.enc_depth}L")


# ── 9. Summary table ──────────────────────────────────────────
print("\n" + "="*60)
print(f"{'Run':<40} {'Final Loss':>10} {'#Params':>10}")
print("="*60)
all_runs = (
    list(zip(labels_mask,  hist_mask,  [m for m,_ in models_mask]))  +
    list(zip(labels_patch, hist_patch, [m for m,_ in models_patch])) +
    list(zip(labels_enc,   hist_enc,   [m for m,_ in models_enc]))
)
for lbl, hist, model in all_runs:
    n_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"{lbl:<40} {hist[-1]:>10.4f} {n_params/1e6:>9.2f}M")
```
